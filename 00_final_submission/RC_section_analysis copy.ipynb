```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import os

# Add the src folder to the Python path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), 'src')))

import mesh
import geometry
```

```python
ReinforcedConcrete = geometry.ReinforcedConcreteColumn(width  = 300, 
                                                       height = 300, 
                                                       concrete_cover = 30,
                                                       rebar_diameter = 30,
                                                       rebar_spacing  = 50)
Mesh = mesh.Mesh(ReinforcedConcrete, "triangle", 25)

Mesh.plot()
plt.savefig("rc.png", bbox_inches='tight', dpi=300)
plt.show()
```

```python
HEB_300 = geometry.H_beam(web_width     =  11.0,
                 web_height    = 262.0,
                 flange_width  = 300.0,
                 flange_height =  19.0)

# Create the mesh of the column
column = mesh.Mesh(HEB_300, mesh_type="quadrilateral", mesh_size=20)

# Print and plot the mesh of the column
column.plot()
plt.savefig("heb.png", bbox_inches='tight', dpi=300)
plt.show()
```

```python
L = geometry.L_Beam(100, 75, 8, 8, 5, 12.95)
Mesh = mesh.Mesh(L, "triangle", 3)

Mesh.plot()
plt.legend()
plt.savefig("L.png", bbox_inches='tight', dpi=300)
plt.show()
```
