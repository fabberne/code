```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']
plt.rc('text', usetex=True)
# set dpi to 300 for better quality
# Material definition
steel = material.Steel_S235()

# Data
strains = np.linspace(-0.008, 0.008, 50000)
stresses = steel.get_stress_vectorized(strains)
e_modul = steel.get_tangent_vectorized(strains)

# Create the main axis
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(12, 3))

# Plot stress-strain on the primary y-axis
ax1.plot(strains * 100, stresses, label=r'$\sigma$ [MPa]')
ax1.set_xlabel(r'strain $\varepsilon$ [$\%$]')
ax1.set_ylabel(r'$\sigma$ [$MPa$], $E$ [$GPa$]')
ax1.set_ylim(-290, 290)
ax1.grid(True, alpha=0.5)

ax1.plot(strains * 100, e_modul / 1000, label=r'$E$ [GPa]', color="C1")
ax1.annotate(r'S235', xy=(0.85, 0.05), xycoords='axes fraction',
			fontsize=12, color='k', ha='left', va='center')
ax1.legend(loc='upper left')



steel = material.Rebar_B500B()

strains = np.linspace(-0.02, 0.02, 50000)
stresses = steel.get_stress_vectorized(strains)
e_modul = steel.get_tangent_vectorized(strains)

ax2.plot(strains * 100, stresses, label=r'$\sigma$ [MPa]')
ax2.set_xlabel(r'strain $\varepsilon$ [$\%$]')
ax2.set_ylim(-580, 580)
ax2.grid(True, alpha=0.5)

ax2.plot(strains * 100, e_modul / 1000, label=r'$E$ [GPa]', color="C1")
ax2.annotate(r'B500B', xy=(0.80, 0.05), xycoords='axes fraction',
			fontsize=12, color='k', ha='left', va='center')
ax2.legend(loc='upper left')


concrete = material.Concrete()

# Data
strains = np.linspace(-0.0003, 0.011, 500)
stresses = concrete.get_stress_vectorized(strains)
e_modul = concrete.get_tangent_vectorized(strains)

ax3.plot(strains * 100, stresses, label=r'$\sigma$ [MPa]')
ax3.set_xlabel(r'strain $\varepsilon$ [$\%$]')
ax3.set_ylim(-4, 37)
ax3.grid(True, alpha=0.5)

ax3.plot(strains * 100, e_modul / 1000, label=r'$E$ [GPa]', color="C1")
ax3.annotate(r'C30/73', xy=(0.77, 0.15), xycoords='axes fraction',
			fontsize=12, color='k', ha='left', va='center')
ax3.legend(loc='upper right')

# Optional: combine legends
lines1, labels1 = ax1.get_legend_handles_labels()
plt.savefig("steel.png", bbox_inches='tight', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']
plt.rc('text', usetex=True)
# set dpi to 300 for better quality
plt.rcParams['figure.dpi'] = 300
# Material definition
steel = material.Rebar_B500B()

# Data
strains = np.linspace(-0.02, 0.02, 50000)
stresses = steel.get_stress_vectorized(strains)
e_modul = steel.get_tangent_vectorized(strains)

# Create the main axis
fig, ax1 = plt.subplots(figsize=(5, 4))

# Plot stress-strain on the primary y-axis
ax1.plot(strains * 100, stresses, label=r'$\sigma$ [MPa]')
ax1.set_xlabel(r'strain $\varepsilon$ [$\%$]')
ax1.set_ylabel(r'stress $\sigma$ [$MPa$]')
ax1.set_ylim(-580, 580)
ax1.grid(True)

# Create the secondary y-axis for Young's modulus
ax2 = ax1.twinx()
ax2.plot(strains * 100, e_modul / 1000, label=r'$E$ [GPa]', color='orange')
ax2.set_ylabel(r'Young\'s modulus $E$ [$GPa$]')
ax2.set_ylim(-290, 290)

# Optional: combine legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.show()
```
