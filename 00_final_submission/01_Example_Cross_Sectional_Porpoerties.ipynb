```python
H    = H_beam(web_width     =   6.5,
              web_height    = 170.0,
              flange_width  = 200.0,
              flange_height =  10.0)

mesh = Mesh(geometry  = H, 
            mesh_type = "quadrilateral", 
            mesh_size = 5)

mesh.plot()
mesh.print()
```

```python
points = [( 0.0, 0.0), 
          ( 5.0, 0.0),
          ( 5.0, 0.75), 
          ( 0.75, 0.75), 
          ( 0.575, 0.825), 
          ( 0.5, 1.0),
          ( 0.5, 9.0),
          ( 0.575, 9.175),
          ( 0.75, 9.25),
          ( 5.0, 9.25),
          ( 5.0,10.0),
          ( 0.0,10.0)]

poly = Polygon(points)
mesh = Mesh(poly, "quadrilateral", 0.25)

mesh.plot()
mesh.print()
```
