```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import os

# Add the src folder to the Python path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), 'src')))

import material
```

```python
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']
# enable latex formatting in matplotlib
plt.rc('text', usetex=True)

steel = material.Steel_S235()

strains = np.linspace(-0.008, 0.008, 500)
stresses = steel.get_stress_vectorized(strains)
e_modul = steel.get_tangent_vectorized(strains)

plt.figure(figsize=(5, 4))
plt.plot(strains * 100, stresses)
plt.plot(strains * 100, e_modul/1000)
plt.xlabel(r'strain $\varepsilon$ [$\%$]')
plt.ylabel(r'stress $\sigma$ [$MPa$]')
plt.title("Steel S235")
plt.grid()
plt.show()
```
