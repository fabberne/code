```python
import sys
import os

# Add the src folder to the Python path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), 'src')))

import numpy as np
import matplotlib.pyplot as plt

from mesh import *
from geometry import *
import structure
import solver
```

```python
HEB_300 = H_beam(web_width     =  11.0,
                 web_height    = 262.0,
                 flange_width  = 300.0,
                 flange_height =  19.0)
beam = Mesh(HEB_300, mesh_type="quadrilateral", mesh_size=100)
beam.plot()
```

```python
struc_3  = structure.Cantilever(beam        = beam, 
								length      = 1000, 
								load_DOFs   = [6+2],
								nodal_loads = [1],
								number_of_sections_per_elemnt = 3)
struc_4  = structure.Cantilever(beam        = beam, 
								length      = 1000, 
								load_DOFs   = [6+2],
								nodal_loads = [1],
								number_of_sections_per_elemnt = 4)
struc_6  = structure.Cantilever(beam        = beam, 
								length      = 1000, 
								load_DOFs   = [6+2],
								nodal_loads = [1],
								number_of_sections_per_elemnt = 6)
struc_10 = structure.Cantilever(beam        = beam, 
								length      = 1000, 
								load_DOFs   = [6+2],
								nodal_loads = [1],
								number_of_sections_per_elemnt = 10)
struc_20 = structure.Cantilever(beam        = beam, 
								length      = 1000, 
								load_DOFs   = [6+2],
								nodal_loads = [1],
								number_of_sections_per_elemnt = 20)
```

```python
non_linear_solver_3  = solver.Nonlinear(struc_3,  constraint="Arc")
non_linear_solver_4  = solver.Nonlinear(struc_4,  constraint="Arc")
non_linear_solver_6  = solver.Nonlinear(struc_6,  constraint="Arc")
non_linear_solver_10 = solver.Nonlinear(struc_10, constraint="Arc")
non_linear_solver_20 = solver.Nonlinear(struc_20, constraint="Arc")
```

```python
increments = np.zeros(200)
increments.fill(1750)

from contextlib import redirect_stdout
with open(os.devnull, 'w') as f, redirect_stdout(f):
	u_history_3 , lambda_history_3  = non_linear_solver_3.solve(increments)
	u_history_4 , lambda_history_4  = non_linear_solver_4.solve(increments)
	u_history_6 , lambda_history_6  = non_linear_solver_6.solve(increments)
	u_history_10, lambda_history_10 = non_linear_solver_10.solve(increments)
	u_history_20, lambda_history_20 = non_linear_solver_20.solve(increments)
```

```python
plt.plot(u_history_3[ :,6+2], lambda_history_3,  label= "3 Sections")
plt.plot(u_history_4[ :,6+2], lambda_history_4,  label= "4 Sections")
plt.plot(u_history_6[ :,6+2], lambda_history_6,  label= "6 Sections")
plt.plot(u_history_10[:,6+2], lambda_history_10, label="10 Sections")
plt.plot(u_history_20[:,6+2], lambda_history_20, label="20 Sections")
plt.legend()
plt.xlim(-1,21) 
```
